```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df.drop(df[['fleet_id','days_arch','days']],axis=1,inplace=True)
df.head()
```

```python
print("\nSummary Statistics of Numeric Columns:")
print(df.describe())
```

```python
df['flag_id'].value_counts().head()
```

```python
flags_of_interest = ['FJ', 'FM', 'KI', 'MH', 'NR', 'PG', 'PW', 'SB']

filtered_df = df[df['flag_id'].isin(flags_of_interest)]
catch_counts = filtered_df['flag_id'].value_counts()
print(catch_counts)
```

```python
# Create a bar plot with custom colors
plt.figure(figsize=(12, 8))
sns.countplot(data=filtered_df, x='flag_id', order=filtered_df['flag_id'].value_counts().index, palette='coolwarm')

# Add labels and title
plt.xlabel('PNA Flag ID')
plt.ylabel('Count')
plt.title('PNA Catch')

# Show plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
gear_catch = df.groupby('gr_id').sum()[['alb_c', 'bet_c', 'blm_c', 'bum_c', 'mls_c', 'skj_c', 'swo_c', 'yft_c']]
print("\nTotal Tuna Catches by Gear Type:")
gear_catch.head()
```

```python
gear_catch.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Total Tuna Catches by Gear Type')
plt.xlabel('Gear Type')
plt.ylabel('Total Catch (mt)')
plt.xticks(rotation=0)
plt.legend(title='Tuna Species')
plt.show()
```

```python
g = sns.FacetGrid(df, col="ez_id", hue="ez_id", col_wrap=4, height=4)
g.map(sns.lineplot, "yr", "bet_c")
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Bigeye Tuna Catch Over Time for Different EEZs')
plt.show()
```
